```python
import plotly.express as px

actual_sales = df[df['SCENARIO'] == 'actual']

# 1. Distribución de ventas realizadas por país
country_sales = actual_sales.groupby('COUNTRY')['AMOUNT'].sum().reset_index()
fig_country_sales = px.bar(country_sales, x='COUNTRY', y='AMOUNT', title="Distribución de Ventas Realizadas por País")
fig_country_sales.update_layout(xaxis_title="País", yaxis_title="Ventas (Monto Total)")

# 2. Distribución de ventas realizadas por mes y año
monthly_sales = actual_sales.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().reset_index()
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['YEAR', 'MONTH']].assign(day=1))
fig_monthly_sales = px.line(monthly_sales, x='Date', y='AMOUNT', title="Distribución de Ventas Realizadas por Mes y Año")
fig_monthly_sales.update_layout(xaxis_title="Fecha (Año, Mes)", yaxis_title="Ventas (Monto Total)")

# 3. Distribución de ventas realizadas por marca
subbrand_sales = actual_sales.groupby('SUBBRAND')['AMOUNT'].sum().reset_index()
fig_subbrand_sales = px.bar(subbrand_sales, x='SUBBRAND', y='AMOUNT', title="Distribución de Ventas Realizadas por Marca (Sub-brand)")
fig_subbrand_sales.update_layout(xaxis_title="Marca (Sub-brand)", yaxis_title="Ventas (Monto Total)")

fig_country_sales.show()
fig_monthly_sales.show()
fig_subbrand_sales.show()
```

## 1. Distribución de Ventas Realizadas por País
Este gráfico muestra cómo se distribuyen las ventas entre los diferentes países, permitiendo identificar claramente dónde se concentran las ventas más altas y cuáles son los países con menor volumen de las mismas.
A simple vista se puede concluir que el país con mayor volumen de ventas es Gran Bretaña (UK) y el país con el menor volumen es España
## 2. Distribución de Ventas Realizadas por Mes y Año
Este gráfico muestra el comportamiento temporal de las ventas, observando patrones tanto anuales como mensuales. Identificar los meses con picos y caídas en ventas ayuda a comprender si hay estacionalidad, como ventas más altas en verano o invierno, y permite realizar predicciones a futuro con mayor precisión.
Lo que se puede ver es que las tendencias son positivas en general, sin embargo, hay una caída de ventas en torno a Enero (que es la tendencia negativa más significativa)
## 3. Distribución de Ventas Realizadas por Marca (Sub-brand)
Este gráfico ayuda a comparar el rendimiento de cada marca o sub-brand, mostrando claramente cuáles son las más vendidas y cuáles tienen menos éxito.
El gráfico viene liderado por Pepsi a la cabeza con una diferencia reseñable respecto a las demás

```python

import plotly.express as px

# Suponiendo que tienes el DataFrame `actual_sales` listo y filtrado

# Encontrar el país con menos ventas y la marca con más ventas en el conjunto de datos de ventas reales
country_least_sales = actual_sales.groupby('COUNTRY')['AMOUNT'].sum().idxmin()
subbrand_most_sales = actual_sales.groupby('SUBBRAND')['AMOUNT'].sum().idxmax()

# Filtrar datos para el país con menos ventas y la marca con más ventas
country_least_sales_data = actual_sales[actual_sales['COUNTRY'] == country_least_sales]
subbrand_most_sales_data = actual_sales[actual_sales['SUBBRAND'] == subbrand_most_sales]

# Agregar los datos por mes y año para analizar tendencia y estacionalidad
country_least_sales_monthly = country_least_sales_data.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().reset_index()
subbrand_most_sales_monthly = subbrand_most_sales_data.groupby(['YEAR', 'MONTH'])['AMOUNT'].sum().reset_index()

# Crear la columna 'Date' para ambos DataFrames
country_least_sales_monthly['Date'] = pd.to_datetime(country_least_sales_monthly[['YEAR', 'MONTH']].assign(day=1))
subbrand_most_sales_monthly['Date'] = pd.to_datetime(subbrand_most_sales_monthly[['YEAR', 'MONTH']].assign(day=1))

# Graficar para el país con menos ventas
fig_country_least_sales = px.line(
    country_least_sales_monthly,
    x='Date', y='AMOUNT',
    title=f"Tendencia y Estacionalidad de Ventas en {country_least_sales} (País con Menos Ventas)"
)
fig_country_least_sales.update_layout(xaxis_title="Fecha (Año, Mes)", yaxis_title="Ventas (Monto Total)")
fig_country_least_sales.show()

# Graficar para la marca con más ventas
fig_subbrand_most_sales = px.line(
    subbrand_most_sales_monthly,
    x='Date', y='AMOUNT',
    title=f"Tendencia y Estacionalidad de Ventas de {subbrand_most_sales} (Marca con Más Ventas)"
)
fig_subbrand_most_sales.update_layout(xaxis_title="Fecha (Año, Mes)", yaxis_title="Ventas (Monto Total)")
fig_subbrand_most_sales.show()
```

## 1. Tendencia y Estacionalidad de Ventas en el País con Menos Ventas
En este gráfico se muestra cómo evolucionan las ventas a lo largo del tiempo en el país con menor volumen de ventas respecto a las marcas que se presentan, podríamos pensar a simple vista que estas ventas irrisorias en España no tienen mucho sentido, no obstsnte, tenemos que tener en cuenta que las marcas de consumo principal en España como Coca Cola no se reflejan en la base de datos. ## 2. Tendencia y Estacionalidad de Ventas de la Marca con Más Ventas
Este gráfico muestra como fluctua la marca con más ventas, esta no tiene una tendencia clara o directamente podemos decir que no la tiene, en contraste, podemos afirmar que tiene una estacionalidad positiva conforme se acerca a la epoca de verano (meses entre mayo y agosto)

```python
import plotly.express as px

# filtramos solo españa de todo el dataset
spain_data = df[df['COUNTRY'] == 'Spain']

# Separamos operaciones realizadas y predicciones
spain_forecast = spain_data[spain_data['SCENARIO'] == 'AI_forecast']
spain_actual = spain_data[spain_data['SCENARIO'] == 'actual']

# merge entre actuals y forecast
comparison_data = spain_forecast.merge(
    spain_actual,
    on=['YEAR', 'MONTH', 'SUBBRAND'],
    suffixes=('_forecast', '_actual')
)

# Calcular error
comparison_data['error'] = abs(comparison_data['AMOUNT_forecast'] - comparison_data['AMOUNT_actual'])
comparison_data['percentage_error'] = (comparison_data['error'] / comparison_data['AMOUNT_actual']) * 100

# paso previo a mostrar gráfico
monthly_forecast_actual = comparison_data.groupby(['YEAR', 'MONTH']).agg({
    'AMOUNT_forecast': 'sum',
    'AMOUNT_actual': 'sum',
    'error': 'mean',
    'percentage_error': 'mean'
}).reset_index()

# añadir fecha para facil visualización
monthly_forecast_actual['Date'] = pd.to_datetime(monthly_forecast_actual[['YEAR', 'MONTH']].assign(day=1))

# 1. Actuals vs predicciones
fig_forecast_actual = px.line(
    monthly_forecast_actual,
    x='Date',
    y=['AMOUNT_forecast', 'AMOUNT_actual'],
    title="Comparación de Predicciones y Ventas Reales en España",
    labels={'value': 'Ventas (Monto Total)', 'variable': 'Tipo'}
)
fig_forecast_actual.update_layout(xaxis_title="Fecha", yaxis_title="Ventas (Monto Total)")
fig_forecast_actual.show()

# 2. Error porcentual y error absoluto
fig_error = px.line(
    monthly_forecast_actual,
    x='Date',
    y='error',
    title="Error Absoluto Medio por Mes en Predicciones en España",
    labels={'error': 'Error Absoluto'}
)
fig_error.update_layout(xaxis_title="Fecha", yaxis_title="Error Absoluto")
fig_error.show()

fig_percentage_error = px.line(
    monthly_forecast_actual,
    x='Date',
    y='percentage_error',
    title="Error Porcentual Medio por Mes en Predicciones en España",
    labels={'percentage_error': 'Error Porcentual (%)'}
)
fig_percentage_error.update_layout(xaxis_title="Fecha", yaxis_title="Error Porcentual (%)")
fig_percentage_error.show()
```
